```python
import pandas as pd
from ast import literal_eval
```

```python
big_data = pd.read_csv("paper_info_1990.csv")
stat_data = pd.read_csv("stat_paper.csv")
```

```python
all_paper = list(big_data.loc[:, 'id'])
all_stat_paper = list(stat_data.loc[:, 'id'])
```

```python
all_data = big_data.set_index('id')
all_data.loc[:,'refereneces'] = all_data.loc[:,'refereneces'].apply(lambda x: literal_eval(x))
```

```python
def dfs_paths(this_id, max_depth, this_depth):
    if this_depth > max_depth:
        return []
    
    # if not found in big data, don't include it and return
    if this_id not in all_paper:
        return []
    
    # if not found in stat data, include it and return
    if this_id not in all_stat_paper:
        return [this_id]
    
    this_references = all_data.loc[this_id, 'refereneces']
    all_this_id = []
    for ref in this_references:
        retval = dfs_paths(ref, max_depth, this_depth+1)
        if len(retval) == 0:
            to_append = [this_id]
            if to_append not in all_this_id:
                all_this_id.append(to_append)
        else:
            for r in retval:
                to_append = []
                if type(r) == str:
                    to_append = [this_id] + [r]
                else:
                    to_append = [this_id] + r
                    
                if to_append not in all_this_id:
                    all_this_id.append(to_append)
            
    return all_this_id
```

```python
result = dfs_paths('53e997a6b7602d9701f78af9', 10, 1)
```
